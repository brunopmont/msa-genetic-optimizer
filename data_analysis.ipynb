```python
"""
data_analysis.ipynb
Analysis and plotting script for evaluating Genetic Algorithm (GA) 
performance against benchmark metrics across BAliBase 3 instances.
"""

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style for publication-ready academic figures
sns.set_theme(style="whitegrid")
plt.rcParams.update({
    "font.size": 12,
    "axes.labelsize": 14,
    "axes.titlesize": 14,
    "xtick.labelsize": 11,
    "ytick.labelsize": 11
})

# =====================================================================
# 1. Load and Process Benchmark Results
# =====================================================================
# Load the aggregated benchmark results generated by the test suite
results_path = "resultados/resultados_gerais_benchmark.txt"

# Alternatively, define the DataFrame directly from experimental data logs
data = {
    "Instance": ["BB11001", "BB11010", "BB12012", "BB12022", "BB20030", 
                 "BB30016", "BB20010", "BB30024", "BB40011", "BB50014"],
    "GA_QScore": [0.988, 0.387, 0.766, 0.992, 0.892, 0.352, 0.908, 0.889, 0.801, 0.929],
    "ACO_QScore": [0.969, 0.299, 0.564, 0.783, 0.895, 0.414, 0.852, 0.726, 0.363, 0.811],
    "TimeSeconds": [0.19, 0.68, 0.70, 7.72, 114.0, 6120.0, 8640.0, 59760.0, 28800.0, 12600.0]
}

df = pd.DataFrame(data)

# Calculate relative performance improvement
df["Improvement_Relative"] = ((df["GA_QScore"] - df["ACO_QScore"]) / df["ACO_QScore"]) * 100
print("--- Benchmark Summary Statistics ---")
print(df.describe())

# =====================================================================
# 2. Performance Comparison Plot (GA vs. ACO Q-Score)
# =====================================================================
plt.figure(figsize=(12, 6))
bar_width = 0.35
indices = range(len(df["Instance"]))

plt.bar([i - bar_width/2 for i in indices], df["GA_QScore"], width=bar_width, label="Proposed GA (Warm Start)", color="#2b5c8f")
plt.bar([i + bar_width/2 for i in indices], df["ACO_QScore"], width=bar_width, label="Reference ACO Literature", color="#d95f02")

plt.xlabel("BAliBase 3 Instances")
plt.ylabel("Q-Score (Alignment Quality)")
plt.title("Comparative Quality Analysis: Genetic Algorithm vs. Ant Colony Optimization")
plt.xticks(ticks=indices, labels=df["Instance"], rotation=45)
plt.ylim(0, 1.1)
plt.legend()
plt.tight_layout()

# Save convergence/comparison figure
plt.savefig("resultados/grafico_convergencia_AG.png", dpi=300)
plt.show()

# =====================================================================
# 3. Computational Scaling Analysis (Curse of Dimensionality)
# =================5====================================================
plt.figure(figsize=(10, 5))
plt.plot(df["Instance"], df["TimeSeconds"] / 3600, marker="o", linestyle="-", color="#2b5c8f", linewidth=2)
plt.xlabel("BAliBase 3 Instances")
plt.ylabel("Execution Time (Hours)")
plt.title("Computational Time Profile & Scaling Behavior")
plt.xticks(rotation=45)
plt.yscale("log")
plt.tight_layout()

plt.savefig("resultados/grafico_convergencia.png", dpi=300)
plt.show()
```
